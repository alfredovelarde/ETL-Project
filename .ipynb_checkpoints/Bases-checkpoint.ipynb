```python
#Importamos librerias para leer la informacion del cilma y manipularlas con pandas
import json
import requests 
import pandas as pd
```

```python
#De la base de datos NOOA solo obtuvimos informacion de la presipitación y temperatura de la ciudad de mexico
#Creamos un arreglo solo con la información que podemos obtener.
#Hacemos una llamda a la pagina por cada tipo de caracteristica del clima y lo apendamos los datos.
getdata=['PRCP','TAVG','TMAX','TMIN']
datatype=[]
for value in getdata:
    r=requests.get("https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GHCND&limit=1000&stationid=GHCND:MXM00076680&units=standard&startdate=2019-01-01&enddate=2019-12-31&datatypeid="+value, headers={"token":"VwDkjYpjrlabCLXsfcxtPBFclLehUEzr"}).json()    
    datatype.append(r['results'])
```

```python
#Obtenemos un muestra de la información
datatype[0][1]
```

```python
#Guradamos cada uno de los datos en listas distintas
prcp_data=[]
prcp_data_date=[]
tavg_data=[]
tavg_data_date=[]
tmin_data=[]
tmin_data_date=[]
tmax_data=[]
tmax_data_date=[]
for each_list in datatype:
     for data in each_list:
            if data['datatype']=='PRCP':
                prcp_data_date.append(data['date'])
                prcp_data.append(data['value'])            
            elif data['datatype']=='TAVG':
                tavg_data_date.append(data['date'])
                tavg_data.append(data['value'])                  
            elif data['datatype']=='TMIN':
                tmin_data_date.append(data['date'])
                tmin_data.append(data['value'])                
            elif data['datatype']=='TMAX':
                tmax_data_date.append(data['date'])
                tmax_data.append(data['value']) 
```

```python
#Utilizamos la libreria datetime para crear fechas y una columna que contiene las fechas de todo 2019
from datetime import datetime, timedelta

inicio = datetime(2019,1,1)
fin    = datetime(2019,12,31)

lista_fechas = [(inicio + timedelta(days=d)).strftime("%Y-%m-%d")
                    for d in range((fin - inicio).days + 1)] 
#print(lista_fechas)

data_df=pd.DataFrame(lista_fechas,columns=['Date'])
data_df.head()
```

```python
#Por cada una de las variables del clima creamos un dataframe.
data_PRCP=pd.DataFrame(prcp_data_date,columns=['Date_PRCP'])
data_PRCP['Precipitation']=pd.DataFrame(prcp_data)
data_TAVG=pd.DataFrame(tavg_data_date,columns=['Date_TempAvg'])
data_TAVG['TempAvg']=pd.DataFrame(tavg_data)
data_TMIN=pd.DataFrame(tmin_data_date,columns=['Date_TempMin'])
data_TMIN['TempMin']=pd.DataFrame(tmin_data)
data_TMAX=pd.DataFrame(tmax_data_date,columns=['Date_TempMax'])
data_TMAX['TempMax']=pd.DataFrame(tmax_data)
```

```python
#Convertimos todas las fechas a date
data_PRCP['Date_PRCP']=pd.to_datetime(data_PRCP['Date_PRCP'])
data_TAVG['Date_TempAvg']=pd.to_datetime(data_TAVG['Date_TempAvg'])
data_TMIN['Date_TempMin']=pd.to_datetime(data_TMIN['Date_TempMin'])
data_TMAX['Date_TempMax']=pd.to_datetime(data_TMAX['Date_TempMax'])
data_df['Date']=pd.to_datetime(data_df['Date'])
```

```python
#Hacemos un join por cada uno de los dataframe para quedarnos con uno solo.
data_df_all=data_df.merge(data_PRCP, left_on='Date', right_on='Date_PRCP', how='left')
data_df_all=data_df_all.merge(data_TAVG, left_on='Date', right_on='Date_TempAvg', how='left')
data_df_all=data_df_all.merge(data_TMIN, left_on='Date', right_on='Date_TempMin', how='left')
data_df_all=data_df_all.merge(data_TMAX, left_on='Date', right_on='Date_TempMax', how='left')
data_df_all[['Date','Precipitation','TempAvg','TempMin','TempMax']]
```

```python
#Leemos los archivos de la calidad del aire
file1 = "Calidad_Aire/2019CO_Monoxido.csv"
file2 = "Calidad_Aire/radiacion_2019UVA.csv"
file3 = "Calidad_Aire/radiacion_2019UVB.csv"
file_one_df = pd.read_csv(file1)
file_two_df = pd.read_csv(file2)
file_three_df = pd.read_csv(file3)
```

```python
#Cambiamos los valores -99 por cero
file_one_df=file_one_df.replace(-99,0)
file_two_df=file_two_df.replace(-99,0)
file_three_df=file_three_df.replace(-99,0)
```

```python
#Agrupamos las bases de datos por fecha 
file_one_df=file_one_df[['FECHA', 'ACO', 'AJM', 'ATI', 'BJU', 'CAM', 'CCA', 'CHO', 'CUA',
       'FAC', 'FAR', 'HGM', 'INN', 'IZT', 'LLA', 'LPR', 'MER', 'MGH', 'MON',
       'MPA', 'NEZ', 'PED', 'SAG', 'SAC', 'SFE', 'SJA', 'TAH', 'TLA', 'TLI',
       'UAX', 'UIZ', 'VIF', 'XAL']].groupby('FECHA').sum()
#
```

```python
file_two_df = file_two_df[['FECHA', 'CHO', 'CUT', 'FAC', 'LAA', 'MER', 'MON', 'MPA', 'PED',
       'SAG', 'SFE', 'TLA']].groupby('FECHA').sum()
```

```python

file_three_df = file_three_df[['FECHA', 'CHO', 'CUT', 'FAC', 'LAA', 'MER', 'MON', 'MPA', 'PED',
       'SAG', 'SFE', 'TLA']].groupby('FECHA').sum()
```

```python
#Calculamos la media de por día en 2019
co_df=file_one_df.mean(axis=1)
co_df.to_frame()
co_df=co_df.reset_index()
uva_df=file_two_df.mean(axis=1)
uva_df.to_frame()
uva_df=uva_df.reset_index()
uvb_df=file_three_df.mean(axis=1)
uvb_df.to_frame()
uvb_df=uva_df.reset_index()
#Convetimos la columna de FECHA en date
co_df['FECHA']=pd.to_datetime(co_df['FECHA'])
uva_df['FECHA']=pd.to_datetime(uva_df['FECHA'])
uvb_df['FECHA']=pd.to_datetime(uvb_df['FECHA'])
```

```python
#Renombramos las columnas
co_df=co_df.rename(columns={0: "n_CO"})
uva_df=uva_df.rename(columns={0: "n_UVA"})
uvb_df=uvb_df[['FECHA',0]]
uvb_df=uvb_df.rename(columns={0: "n_UVB"})
```

```python
#Con la libreria glob importamos todos los archivos de la carpeta "Datos_Ecobici"
import glob
csv_files = glob.glob('Datos_Ecobici/*.csv')
# Mostrar el archivo csv_files, el cual es una lista de nombres
eco_bici = pd.DataFrame(columns=('Genero_Usuario', 'Edad_Usuario', 'Bici', 'Ciclo_Estacion_Retiro',
       'Fecha_Retiro', 'Hora_Retiro', 'Ciclo_Estacion_Arribo', 'Fecha_Arribo',
       'Hora_Arribo'))
csv_files
```

```python
# Escribimos un loop que irá a través de cada uno de los nombres de archivo a través de globbing y el resultado final será la lista dataframes
lista_data=[]
for filename in csv_files:
    data = pd.read_csv(filename)
    lista_data.append(data)
lista_data
```

```python
#Apendamos todos los dataframe en uno solo
df = eco_bici.append([lista_data[0], lista_data[1],lista_data[2],lista_data[3],lista_data[4],lista_data[5],lista_data[6],lista_data[7],lista_data[8]])
```

```python
#Solo escogemos las columnas que necesitamos
df=df[['Bici','Edad_Usuario','Fecha_Retiro']]
df['Edad_Usuario']=df['Edad_Usuario'].astype(str).astype(int)
df['Edad_Usuario']
```

```python
#Agrupamos la información por fecha y contamos el número de bicis y la edad promedio del pasajero.
df=df.groupby('Fecha_Retiro').agg({'Bici':'count', 'Edad_Usuario':'mean'})
```

```python
#Reset del index y convertimos la columna fecha a date
df=df.reset_index()
df=df[['Fecha_Retiro','Bici','Edad_Usuario']]
df['Fecha_Retiro']=pd.to_datetime(df['Fecha_Retiro'])
```

```python
#Combinamos la informacion de todas nuestras bases de datos "Clima","Ecobici","Contaminacion del aire"
postgres_database=pd.merge(data_df_all,df,how = 'left',left_on="Date",right_on="Fecha_Retiro")
postgres_database=pd.merge(postgres_database,co_df,how = 'left',left_on="Date",right_on="FECHA")
postgres_database=pd.merge(postgres_database,uva_df,how = 'left',left_on="Date",right_on="FECHA")
postgres_database=pd.merge(postgres_database,uvb_df,how = 'left',left_on="Date",right_on="FECHA")
```

```python
#Seleccinamos solo las columnas que queremos mantener
postgres_database=postgres_database[['Date', 'Precipitation', 'TempAvg', 'TempMin', 'TempMax',
       'Bici', 'Edad_Usuario', 'n_CO', 'n_UVA',
       'n_UVB']]
```

```python
postgres_database.columns
```

```python
#Hacemos la conexion a la base de datos 
from sqlalchemy import create_engine
connection_string = "postgres:postgres@localhost:5432/ETL_Project"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#Creamos una copia de nuestro dataframe
ecobici_transformed = postgres_database.copy()
```

```python
#Colocamos la fecha como nuestro index
ecobici_transformed.set_index("Date", inplace=True)

ecobici_transformed.head()
```

```python
#Obtenemos el engine de nuestras tablas
engine.table_names()
```

```python
#Agregamos nuestro dataframe a sqlpostgress
ecobici_transformed.to_sql(name='ecobici', con=engine, if_exists='append', index=True)
```
